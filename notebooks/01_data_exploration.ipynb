```python
import sys
import os

sys.path.append("..")
os.chdir("..")

import pandas as pd

train_df = pd.read_csv("data/train/_annotations.csv")
train_df.head()
```

```python
train_df["class"].value_counts()
```

```python
print("Total annotation rows:", len(train_df))
print("Unique images:", train_df["filename"].nunique())
```

```python
classes_per_image = train_df.groupby("filename")["class"].nunique()
classes_per_image.value_counts().sort_index()
```

```python
image_classes = train_df.groupby("filename")["class"].apply(set)

for cls in ["Cavity", "Fillings", "Impacted Tooth", "Implant"]:
    count = image_classes.apply(lambda s: cls in s).sum()
    pct = 100 * count / len(image_classes)
    print(f"{cls}: {count} images ({pct:.1f}%)")
```
